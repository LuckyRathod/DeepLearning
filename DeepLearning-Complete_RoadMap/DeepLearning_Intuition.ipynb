## T2 - How does Neural Network work ?

Neural Network Architecture

Suppose we have 3 features (x1,x2,x3) which will be first passed through Input layer to Hidden Layer where neuron is present.
There are some weights (w1,w2,w3) assigned at Line between Input layer and Hidden layer neuron.These weight will also get passed to hidden neuron along with inputs (x1,x2,x3).

Two types of Operations are performed in Hidden Neuron
1. Summation of Weights and Inputs + Bias
2. Activation function is appplied to output of step 1

Sigmoid Activation Function - 1/(1+e^-y) - It will transform value between 0 to 1

If it is 0.5 then that particular neuron will not get activated , Otherwise it will get activated

Example - If i keep a hot object on my hand , Weights will become high - Activation function will be applied and these Particular neuron in my that hand will get activated . Other hand neurons will not get activated.

So only that hand neuron information will be passed to brain where the hot object is placed , Based on that brain will react what to do further.

These 2 step process is applied to each and every neuron in Hidden layer.

Output layer neuron also contains Activation function.



## T3 - Activation Functions - Part 1

##### 1. Sigmoid Activation Function

Activation funtion in Neuron is applied after Summation of Weight and inputs + Bias .Sigmoid Activation function transforms value between 0 to 1

y = Summation of weights and inputs + bias
Sigmoid Activation Function - 1/(1+e^-y)

Why Activation Function is Needed ?

Example - We need some function to activated the neurons which are connected to brain which will give output.Whenever any hot object is placed on my one hand . Neurons of that hand will get activated (>0.5 value) with the help of Activation function . Which will then send the signal to Brain.Neurons of other hand will not get activated becasue its value is 0 and no hot object is placed on that.



##### 2. RELU Activation Function

y =  Summation of Weights and Inputs + Bias 

RELU Activation function will find maximum between value of y and 0.

I-E If my y value is negative it will be transformed to 0 , otherwise 1



Classification problems - Hidden layer -RELU , Output layer - Sigmoid
Regression problem - RELU

### Forward Propogation

All my input features(x1,x2,x3) will get passed through Input layer to Hidden layer neuron.Neuron process two operations
1. Summation of weights and inputs + bias
2. Sigmoid activation function is applied to y (from step1)



Sigmoid Activation function(Whether neuron is activated or not) output is then passed to output layer (Again Output layer contains activation function) it will give the prediction (y hat).

Suppose yhat= 0

Now in training dataset we already know tha output (y = 1) for these row . So now we will calculate LOSS 

LOSS = (y-yhat)^2

So for above row , our Loss = 1 (Loss value is higher ) . 
Loss value should be reduced to minimal value.(By using optimizer we can reduce the LOSS value)

In order to reduce value BACKPROPOGATION is done (Weights Updation)
We should adjust the weights (w1,w2,w3,w4) in such a way that yhat=1



#### Weigths Updation takes place in following way

We will make use of Learning rate and Derivative of Loss is done w.r.t w4.

Similarly we will update weights of w1,w2,w3



Once we get updated value , We will again do Forward Propogation.These process will get continue for some no of epochs untill the LOSS value is reduced completely
