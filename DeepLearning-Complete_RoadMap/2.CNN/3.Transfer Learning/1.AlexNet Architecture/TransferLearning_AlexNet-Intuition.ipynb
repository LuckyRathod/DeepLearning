### T50 -- Transfer Learning 

#### 1. AlexNet Architecture

Alexnet Architecture was trained on Imagenet dataset.

ImageNet contains more than 20,000 categories with a typical category, such as "balloon" or "strawberry", consisting of several hundred images.



1. Input size of an image in Alexnet Architecture should be of 227*227*3
2. 1st Convolution layer contains : Filter of Size(11*11) , 96 Filters are used with stride = 4

Formula to get an output of 1st convolution layer is (n+2p-f)/s + 1 = No padding = 0 = ((227-11)/4) + 1 = 55

3. 1st Max pooling layer contains : Filters size (3*3) , stride =2 

Output of 1st max pooling layer (n+2p-f)/s +1 = (55-3)/2 +1 = 27

4. 2nd Convolution Layer : Filter (5*5) ,pad=2,256 kernels , by default stride =1

Output of 2nd cnvolution layer (n+2p-f)/s + 1 = (27+4-5)/1 +1 = 27

5. 2nd Max pooling layer : Filter (3*3) , stride =2 

Output of 2nd max pooling layer (27-3)/2 +1 = 13

6. 3rd Convolution layer : Filter(3*3),pad=1 , 384 kernels , stride=1

Output of 3rd Convolution layer = 13

7. 4th Convolution layer : Filter(3*3),pad=1 , 384 kernels , stride=1
Output of 4th convolution layer = 13

8. 5th Covolution layer : Filter (3*3),pad=1,256 kernels 
Output of 5th convolution layer = 13

9. 3rd Max pooling : (3*3),stride=2
Output : 6
