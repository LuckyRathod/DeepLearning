There are 2 variations of VGGNET Architecture
1. VGG16
2. VGG19

Alexnet and VGG uses relu , So even though we create deep neural net , It take cares of Vanishing Gradient Descent problem

#### VGG 16 ----- There are total 16 layers 



Whenever we pass an image in VGG16 , Its image size reduces by half

#### In all Max pooling layers
